```python
# drop Unnamed: 0
df.drop('Unnamed: 0', axis=1, inplace=True)
print(df.shape)
```

```python
# identify_consts
import preppy.utils as utils

consts = utils.functions.identify_consts(df)
consts
```

```python
df.drop(consts, axis=1, inplace=True)
print(df.shape)
```

```python
quasi_consts = utils.functions.identify_quasi_consts(df)
quasi_consts
```

```python
utils.functions.check_row_duplicates(df)
```

```python
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.duplicated.html
df[df.duplicated(subset=['fake_zipcode'])].count()
```

```python
print(df.shape)
df.drop_duplicates(subset=['fake_zipcode'], inplace=True)
print(df.shape)
```

```python
duplicate_features = utils.functions.check_col_duplicates(df)
print(duplicate_features)
```

```python
df.drop(duplicate_features, axis=1, inplace=True)
print(df.shape)
```

```python
df.describe()
```

```python
df = utils.functions.handle_standard_scaler(df)
df.describe()
```

```python
df = utils.functions.handle_minmax_scaler(df)
df.describe()
```

```python
df.isnull().sum()
```

```python
df = utils.functions.handle_missing_values(df)
df.isnull().sum()
```

```python
df = utils.functions.handle_outliers(df)
```
