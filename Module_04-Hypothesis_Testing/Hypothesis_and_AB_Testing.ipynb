## Null and Alternative Hypothesis

* Null hypothesis: Mean 1 = Mean 2
* Alternate hypothesis: Mean 1 $\ne$ Mean 2

**Hypothesis Example**

Is the mean of sample 1 different from the mean of sample 2? In business, AB testing is driven by success. Academics will document differences that may not be important to business.

Please review https://www.statisticshowto.com/probability-and-statistics/hypothesis-testing/one-tailed-test-or-two/

**Operators**

- For a one tailed null hypothesis we see $\leq$ or $\geq$
- For a one tailed alt hypothesis we see $<$ or $>$
- For a two tailed null hypothesis we see the symbol $=$
- For a two tailed alt hypothesis we see the symbol $\ne$

**For a One Tailed Test**
- Null Hypothesis – There is no significant difference between the means of groups
- Alternate Hypothesis – There is a significant difference between the means of groups in that the mean of the group is less or greater than some x

**For a Two Tailed Test**
- Null Hypothesis – There is no significant difference between the means of groups
- Alternate Hypothesis – There is a significant difference between the means of groups either greater than x or less than x

For example,

**For a One Tailed Test**
- a null hypothesis might be $\bar{x}_1 \geq \bar{x}_2$,
- and the alt hypothesis would be $\bar{x}_1 < \bar{x}_2$

**For a Two Tailed Test**
- a null hypothesis might be $\bar{x}_1 = \bar{x}_2$,
- and the alt hypothesis would be $\bar{x}_1 \ne \bar{x}_2$

Remember, we never accept the null hypothesis, we either reject it or fail to reject it.

If our p value is less than the alpha value then the evidence points to rejecting the null hypothesis. If our p values is less than the alpha value then we can say our results are statistcally significant. We found something that is probably not the result of chance. But beware.

The following list clarifies some issues that are commonly misunderstood regarding p-values:

* The p-value is not the probability that the null hypothesis is true, or the probability that the alternative hypothesis is false. A p-value can indicate the degree of compatibility between a dataset and a particular hypothetical explanation (such as a null hypothesis). Specifically, the p-value can be taken as the prior probability of obtaining an effect that is at least as extreme as the observed effect, given that the null hypothesis is true. This should not be confused with the posterior probability that the null hypothesis is true given the observed effect (see prosecutor's fallacy). In fact, frequentist statistics does not attach probabilities to hypotheses.
* The p-value is not the probability that the observed effects were produced by random chance alone. The p-value is computed under the assumption that a certain model, usually the null hypothesis, is true. This means that the p-value is a statement about the relation of the data to that hypothesis.
* The 0.05 significance level is merely a convention. The 0.05 significance level (alpha level) is often used as the boundary between a statistically significant and a statistically non-significant p-value. However, this does not imply that there is generally a scientific reason to consider results on opposite sides of any threshold as qualitatively different.
* The p-value does not indicate the size or importance of the observed effect. A small p-value can be observed for an effect that is not meaningful or important. In fact, the larger the sample size, the smaller the minimum effect needed to produce a statistically significant p-value (see effect size). Visualizing effect sizes is a critical component of a data-analysis method called estimation statistics.

**Explanation of the math behind the t-test:**

The independent t-test (also known as the two-sample t-test) is used to determine if there is a statistically significant difference between the means of two independent groups.

Here's a breakdown of the math involved:

1. **Hypotheses:**
   * **Null Hypothesis (H0):** The means of the two groups are equal.
   * **Alternative Hypothesis (H1):** The means of the two groups are not equal.

2. **Assumptions:**
   * The two samples are independent.
   * The data in each group are normally distributed.
   * The variances of the two groups are equal (homogeneity of variance). If this assumption is violated, you can use Welch's t-test, which is a variation that doesn't assume equal variances.

3. **Calculation of the t-statistic:**

   The t-statistic is calculated as follows:

   $t = (\bar{x}_1 - \bar{x}_2) / \sqrt{\frac{\sigma_1}{n_1} + \frac{\sigma_2}{n_2}}$

   ```
   t = (mean1 - mean2) / sqrt((s1^2/n1) + (s2^2/n2))
   ```

   where:

   * mean1 and mean2 are the sample means of the two groups
   * s1 and s2 are the sample standard deviations of the two groups
   * n1 and n2 are the sample sizes of the two groups

4. **Degrees of Freedom:**

   The degrees of freedom for the t-test are calculated as follows:

   ```
   ddof = n1 + n2 - 2
   ```

5. **P-value:**

   * The calculated t-statistic is compared to the t-distribution with the calculated degrees of freedom.
   * The p-value is the probability of observing a t-statistic as extreme as (or more extreme than) the one calculated, assuming the null hypothesis is true.

6. **Decision:**

   * If the p-value is less than the chosen significance level (alpha, typically 0.05), we reject the null hypothesis and conclude that there is a statistically significant difference between the means of the two groups.
   * If the p-value is greater than or equal to alpha, we fail to reject the null hypothesis and conclude that there is not enough evidence to suggest a significant difference between the means.

**Important Note:**

* This assumes equal variances. If you suspect unequal variances, use `stats.ttest_ind(group1, group2, equal_var=False)` to perform Welch's t-test.

**What are Expected Frequencies?**

In the Chi-Square test of independence, expected frequencies represent the counts we would anticipate in each cell of the contingency table if there were *no association* between the two categorical variables being examined. In other words, they reflect the frequencies we'd expect to see if the null hypothesis (independence) were true.

**How are they Computed?**

The expected frequency for a particular cell is calculated as follows:

```
Expected Frequency = (Row Total * Column Total) / Grand Total
```

where:

* **Row Total:** The sum of all frequencies in the row containing the cell.
* **Column Total:** The sum of all frequencies in the column containing the cell.
* **Grand Total:** The total number of observations in the entire contingency table.

**Intuition**

The formula essentially distributes the total observations proportionally across the cells based on the marginal totals (row and column totals). If the variables were truly independent, we'd expect the observed frequencies to closely align with these expected frequencies. Any significant deviation between the observed and expected frequencies suggests a potential association between the variables.

**Example**

Consider a simplified 2x2 contingency table:

```
        Outcome 1  Outcome 2
Group A     10         20
Group B     15         25
```

* **Row Totals:**
    * Group A: 10 + 20 = 30
    * Group B: 15 + 25 = 40
* **Column Totals:**
    * Outcome 1: 10 + 15 = 25
    * Outcome 2: 20 + 25 = 45
* **Grand Total:** 30 + 40 = 70

Let's calculate the expected frequency for the cell in the first row and first column (Group A, Outcome 1):

```
Expected Frequency = (30 * 25) / 70
                    = 10.71
```

Similarly, you would calculate the expected frequencies for the remaining cells.

**In the Code**

In the Python code you provided, the `stats.chi2_contingency` function automatically calculates these expected frequencies and stores them in the `expected` variable. You can then print or analyze these values to gain insights into the potential association between your variables.

**Key Points**

* Expected frequencies serve as a benchmark for comparison with the observed frequencies.
* Large discrepancies between observed and expected frequencies contribute to a higher Chi-Square statistic, suggesting dependence between the variables.
* The Chi-Square test ultimately assesses whether these discrepancies are statistically significant, i.e., unlikely to have occurred by chance if the variables were truly independent.

### Fisher's Exact Test

Although in practice it is employed when sample sizes are small, it is valid for all sample sizes. It is named after its inventor, Ronald Fisher, and is one of a class of exact tests, so called because the significance of the deviation from a null hypothesis (e.g., P-value) can be calculated exactly, rather than relying on an approximation that becomes exact in the limit as the sample size grows to infinity, as with many statistical tests.

Fisher's exact test is a statistical test used to **determine if there are nonrandom associations between two categorical variables**.

https://mathworld.wolfram.com/FishersExactTest.html

Gemini September 11, 2024

**What is Fisher's Exact Test?**

Fisher's exact test is a statistical significance test used in the analysis of contingency tables. It's particularly useful when dealing with small sample sizes where the assumptions of other tests, like the chi-squared test, might not be met. It assesses whether there's a significant association between two categorical variables.

**When to Use It**

* **Small Sample Sizes:** When the expected frequencies in any of the cells of your contingency table are less than 5, the chi-squared test might not be reliable. Fisher's exact test is appropriate in these cases.
* **2x2 Contingency Tables:** It's primarily designed for 2x2 contingency tables (two rows and two columns), but extensions exist for larger tables.

**How It Works**

1. **Null Hypothesis:** The null hypothesis states that there is no association between the two categorical variables.

2. **Contingency Table:** The data is organized into a contingency table, showing the frequencies of each combination of the two variables.

3. **Calculate Probabilities:** Fisher's exact test calculates the probability of obtaining the observed table or a more extreme table, assuming the null hypothesis is true. It does this by considering all possible tables that could have arisen with the same row and column totals as the observed table.

4. **P-value:** The p-value is the sum of the probabilities of all tables as extreme as or more extreme than the observed table.

5. **Decision:**
   * If the p-value is less than the chosen significance level (alpha, typically 0.05), we reject the null hypothesis and conclude that there is a significant association between the two variables.
   * If the p-value is greater than or equal to alpha, we fail to reject the null hypothesis and conclude that there is not enough evidence to suggest an association.

**Advantages**

* **Accurate for Small Samples:** Provides accurate results even with small sample sizes where other tests might be inaccurate.
* **No Approximations:** Doesn't rely on approximations like the chi-squared test, making it more exact.

**Disadvantages**

* **Computationally Intensive:** Can be computationally intensive for large tables, although modern computers handle most cases efficiently.
* **Limited to 2x2 Tables (Primarily):** While extensions exist for larger tables, it's most straightforward for 2x2 tables.

**Example**

Suppose you're testing whether there's an association between gender (male/female) and preference for a new product (like/dislike). You collect data from a small sample and organize it into a contingency table:

```
             Like  Dislike
Male         8      3
Female       5      6
```

Fisher's exact test would calculate the probability of getting this table or a more extreme one, assuming no association between gender and preference. If the p-value is small, you'd conclude there's evidence of an association.

### Effect Size

In statistics, an effect size is a value measuring the strength of the relationship between two variables in a population, or a sample-based estimate of that quantity. It can refer to the value of a statistic calculated from a sample of data, the value of a parameter for a hypothetical population, or to the equation that operationalizes how statistics or parameters lead to the effect size value.

https://en.wikipedia.org/wiki/Effect_size

https://www.simplypsychology.org/effect-size.html

Gemini, September 11, 2024

**Sample Size**

* **Definition:** The sample size is the number of observations or individuals included in a study or experiment.
* **Importance:** It plays a crucial role in determining the reliability and generalizability of the study's findings.

**Effect Size**

* **Definition:** Effect size is a quantitative measure of the magnitude of the phenomenon being studied. It indicates the strength of the relationship between variables or the difference between groups.
* **Importance:** It provides a standardized way to assess the practical significance of a research finding, independent of sample size.

**Relationship between Sample Size and Effect Size**

1. **Sample Size and Statistical Significance:**

* Larger sample sizes increase the power of a statistical test, making it more likely to detect even small effects as statistically significant.
* With a small sample size, a large effect might not be detected as statistically significant, leading to a Type II error (false negative).

2. **Sample Size and Precision of Effect Size Estimation:**

* Larger sample sizes lead to more precise estimates of the effect size. The larger the sample, the closer the sample estimate is likely to be to the true population effect size.
* Smaller sample sizes result in wider confidence intervals around the effect size estimate, indicating greater uncertainty about the true effect size.

3. **Effect Size is Independent of Sample Size:**

* While sample size influences the ability to detect an effect and the precision of its estimation, the actual effect size itself is a characteristic of the population or phenomenon being studied and is not influenced by the sample size.

**Implications**

* **Planning Research:** Researchers should carefully consider the desired effect size and the level of statistical power when determining the appropriate sample size for their study.
* **Interpreting Research:** When evaluating research findings, it's important to consider both statistical significance (p-value) and effect size. A statistically significant result with a small effect size might not be practically meaningful.
* **Meta-Analysis:** Effect sizes allow researchers to combine results from multiple studies with varying sample sizes in a meta-analysis to obtain a more comprehensive understanding of the phenomenon.

**In summary:**

* Sample size affects the ability to detect an effect and the precision of its estimation.
* Effect size is a measure of the magnitude of the effect and is independent of sample size.
* Both sample size and effect size are important considerations in research design and interpretation.

### One Tail vs Two Tail AB Tests in Terms of CVR

* Both Control and Treatment have equal conversion rates:
    * null $Control (A) = Treatment (B)$
    * alt $Control (A) \ne Treatment (B)$
* Treatment group's conversion rate is no better than the Control group's and could be worse:
    * null $Control (A) \geq Treatment (B)$
    * alt $Control (A) \lt Treatment (B)$
* One tail tests look for an improvement in customer experience

### Lift

Lift indicates if the treatment is better than the control. Lift is a percentage of how the treatment compares to the null hypothesis (control group).

Formula:

### Test of Proportions

$z = p_{sample} - p_{expected} / \sqrt{\frac{p_{expected} (1 - p_{expected})}{n}}$

Gemini, September 11, 2024

**Our Scenario:**

We have a new marketing campaign and want to assess its impact on brand awareness. We surveyed 100 people and found that 35 of them recognize the brand. We want to test if this proportion is significantly different from the previous brand awareness rate of 25%.

**Hypothesis Testing:**

1. **Null Hypothesis (H0):** The new marketing campaign has not increased brand awareness. In other words, the proportion of people who recognize the brand remains at 25%.
2. **Alternative Hypothesis (H1):** The new marketing campaign has increased brand awareness. The proportion of people who recognize the brand is greater than 25%.

**Test Statistic:**

The test statistic we'll use is the z-statistic, which is calculated as follows:

```
z = (p_sample - p_expected) / sqrt(p_expected * (1 - p_expected) / n)
```

where:
* `p_sample` is the sample proportion (35/100 in our case)
* `p_expected` is the hypothesized proportion under the null hypothesis (25% or 0.25)
* `n` is the sample size (100)

**P-value:**

The p-value is the probability of obtaining a z-statistic as extreme or more extreme than the one we calculated, assuming the null hypothesis is true. We can find the p-value using a z-table or statistical software.

**Decision:**

* If the p-value is less than our chosen significance level (usually 0.05), we reject the null hypothesis and conclude that there is evidence to support the claim that the new marketing campaign has increased brand awareness.
* If the p-value is greater than or equal to 0.05, we fail to reject the null hypothesis and conclude that there is not enough evidence to support the claim.

Now, let's calculate the z-statistic and p-value for our scenario:

```
# Calculate the z-statistic
z = ((35/100) - 0.25) / sqrt(0.25 * (1 - 0.25) / 100)
z = 2.00

# Find the p-value (using a one-tailed test since we're testing for an increase)
p_value = 1 - stats.norm.cdf(z)
p_value = 0.0228
```

**Interpretation:**

The z-statistic of 2.00 indicates that the observed proportion (35%) is 2 standard deviations above the expected proportion (25%). The p-value of 0.0228 means there's a 2.28% chance of observing a proportion as extreme or more extreme than 35%, assuming the null hypothesis is true.

Since the p-value (0.0228) is less than the significance level (0.05), we reject the null hypothesis. We can conclude that there is evidence to support the claim that the new marketing campaign has increased brand awareness.
