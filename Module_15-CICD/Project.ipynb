## YML

**What is a YAML file?**

* **YAML** stands for "YAML Ain't Markup Language" or "Yet Another Markup Language" depending who you ask.  It's a way to store information in a format that's both easy for humans to read and easy for computers to understand. Think of it like a simple, organized way to write data.
* **.yml** or **.yaml**  These are the file extensions used for files containing YAML.

**Why use YAML?**

* **Human Readable:**  YAML uses a simple structure with indentation and a few key symbols to represent data. This makes it much cleaner and easier to understand than other formats like XML or JSON.
* **Flexible:** YAML can handle various data types like numbers, text, lists, and even nested structures.
* **Configuration Files:** YAML is super popular for configuration files. Many programs use it to store settings and preferences.
* **Data Serialization:**  It's a handy way to exchange information between different programs, regardless of the programming language they use.
* **Used in DevOps:** Tools like Ansible (for automating tasks) and Kubernetes (for managing containers) rely heavily on YAML files.

**Basic YAML Structure**

YAML relies on a few key elements:

* **Indentation:**  Spaces are used to show the structure of your data.  Consistent indentation is crucial!
* **Key-value pairs:**  Data is often represented as `key: value`. For example:  `name: John Doe`
* **Lists:** Items in a list start with a dash (`-`). For example:
   ```yaml
   - apple
   - banana
   - cherry
   ```

**Example**

```yaml
name: John Doe
age: 30
city: New York
skills:
  - Python
  - Java
  - JavaScript
```

This simple example shows how YAML can store information about a person in a clear and organized way.


This example shows pseudo code for a CICD project.

```
   name: Deploy to Hugging Face Spaces

   on:
     workflow_dispatch:
     push:
       branches: [ main ]

   jobs:
     test:
       runs-on: ubuntu-latest
       steps:
         - name: Checkout repository
           uses: actions/checkout@v4
         - name: Set up Python
           uses: actions/setup-python@v5
           with:
             python-version: '3.13'  # Or your desired version
         - name: Install dependencies
           run: pip install -r requirements.txt
         - name: Run pytest
           run: pytest tests/  # Run tests in the 'tests' directory

     deploy:
       needs: test  # Run tests before deployment
       runs-on: ubuntu-latest
       steps:
         - name: Checkout repository
           uses: actions/checkout@v4
         - name: Install dependencies
           run: pip install -r requirements.txt
         - name: Push to Hugging Face Hub
           uses: huggingface/hub-actions/push@main
           with:
             token: ${{ secrets.HF_TOKEN }}
             repo_id: <your_username>/<your_space_name>
             branch: main
```

**2. Set Up GitHub**

* **Create a repository:** Create a new GitHub repository.
* **Commit and push:** Add and commit all your project files to the repository and push it to GitHub.

**3. Set Up Hugging Face Space**

* **Create a Space:** Create a new Space on Hugging Face and choose the "Gradio" SDK.
* **Link to GitHub:**
   * Go to your Space's settings.
   * In the "Git" or "Repository" section, link your Space to your GitHub repository by providing the repository URL.
   * Choose the branch you want to use (usually `main`).

**4. Configure GitHub Actions**

* **Create a workflow file:** In your GitHub repository, create a `.github/workflows/deploy.yml` file:

   ```yaml
   name: Deploy to Hugging Face Spaces

   on:
     workflow_dispatch:
     push:
       branches: [ main ]

   jobs:
     test:
       runs-on: ubuntu-latest
       steps:
         - name: Checkout repository
           uses: actions/checkout@v3
         - name: Set up Python
           uses: actions/setup-python@v4
           with:
             python-version: '3.10'  # Or your desired version
         - name: Install dependencies
           run: pip install -r requirements.txt
         - name: Run pytest
           run: pytest tests/  # Run tests in the 'tests' directory

     deploy:
       needs: test  # Run tests before deployment
       runs-on: ubuntu-latest
       steps:
         - name: Checkout repository
           uses: actions/checkout@v3
         - name: Install dependencies
           run: pip install -r requirements.txt
         - name: Push to Hugging Face Hub
           uses: huggingface/hub-actions/push@main
           with:
             token: ${{ secrets.HF_TOKEN }}
             repo_id: <your_username>/<your_space_name>
             branch: main
   ```

* **Add `HF_TOKEN` secret:**
   * Go to your Hugging Face profile settings and generate a new access token with write access.
   * In your GitHub repository settings, go to "Secrets and variables" -> "Actions" and create a new repository secret named `HF_TOKEN` with your Hugging Face token.

**5. Explanation**

* **Workflow:** This workflow defines two jobs: "test" and "deploy".
* **Testing:** The "test" job runs your pytest tests.
* **Deployment:** The "deploy" job depends on the "test" job, so it will only run if the tests pass. It then deploys your code and model to your Hugging Face Space.
* **Linking:** Linking your Space to GitHub allows for automatic synchronization and updates whenever you push changes to your repository.

**Important Notes:**

* **Adapt to your needs:** Adjust the code and workflow file to match your specific project structure, model training, and testing requirements.
* **Private Spaces:** If you have a private Space, you'll need to adjust the `repo_id` and potentially use a different type of token in the workflow file.
* **Troubleshooting:** Check the "Actions" tab in your GitHub repository and the logs in your Hugging Face Space for any errors during testing or deployment.

**Explanation of using `pip` for Python installation**

While `pyenv` is a popular tool for managing multiple Python versions, you can use `pip` itself to install specific Python versions within a virtual environment. This method provides flexibility, especially if you prefer to manage everything with `pip`.

**Important Notes**

* **`python` command:** After installing Python 3.1x with `pip`, the `python` command within the virtual environment should point to the newly installed Python 3.1x. You can verify this by running `python --version`.
* **Requirements:**  Make sure to update your `requirements.txt` file to include `gradio` and any other packages you install within the environment.
* **Windows:** The activation step might be slightly different on Windows. Refer to the official Python documentation for Windows-specific instructions.
